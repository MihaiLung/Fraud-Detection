# Logistic regression

## Approach and Considerations

The data is a classic case of an underrepresented interest case - whereas our main priority is making sure we identify correctly all fraudulent transactions while minimising the number of legitimate transactions false positives, a classic regression on the whole data would be a poor way of going about this. 

Classic regression will try to minimise the number of false matches. However, in our data, only 0.17% of observations are fraudulent. Classic regression would be 99.83% right by simply classifying everything as non-fraudulent, which is a likely outcome for the model. Minimising false positives (classifying a legitimate transaction incorrectly as a fraud) and false negatives (classifying a fraudulent transaction incorrectly as legitimate) usually involves a trade-off betwee the two, and therefore without further adjustment the model will end up prioritising minimising false positives much more than false negatives, the opposite of the behaviour we actually want.

The main options for this situation are as follows:
 - re-sample the data - not possible in this case
 - undersample the majority class
 - oversample the minority class
 
We will combine the latter two options from above. We will undersample the majority class by random sampling so the number of observations is comparable to those of the minority class. We will also synthethically enhancethe minority class using SMOTE to help generalise our estimates.

## Oversampling

The obvious alternative to synthetic oversampling is to oversample with replacement. This would consist of drawing a sample from the original minority class sample however replacing each element drawn, as such enabling the new sample to exceed the starting one in size. However, this approach suffers from being exposed to randomness-induced errors:
- there is a small but non-zero chance that the sample drawn will be highly unbalanced or will overrepresent outliers to the point of skewing the results
- if overdone the model might lose generality and what it does might be better described as learning the locations of the minority class sample rather than finding a useful location to generalise its traits

**SMOTE** (synthetic minority oversampling technique) is a method of getting around the lack of generality of sampling with replacement. Rather than oversample by drawing from the same observations, we do it by generating new synthetic observations from random spots on the line space between nearby members of the original minority observations. 

This method is not perfect. It does assume that each cluster of minority observations represents a convex set (although is robust to non-convexity of the minority class as long as it is spread out in relatively convex clusters). Outliers do represent a problem, although not more so than other methods - by looking at the nearest neighbours rather than a random observation the synthetic observations will overwhelmingly be representative of the "correct" data. So if the data is reasonably clean SMOTE should be a good way to expand the data in a manner that does not affect its generality.

## SMOTE

For demonstration purposes, I will implement the SMOTE algorithm myself rather than use a pre-existing package such as imbalanced-learn.

```python
def SMOTE(data,mult,K):
    #data = the training minority class data
    #mult = how many new samples to generate as a proportion of the current data. So,
        #if mult=2 then we generate 2 new observations for each one in the current dataset.
    #K = the number of k-nearest neighbours to find
    
    #Find the K nearest neighbors and the relevant distance for each training data
    KNN=NearestNeighbors(n_neighbors=K+1,algorithm='ball_tree').fit(data)
    KNN_distances, KNN_indices=KNN.kneighbors(data)
    
    #Generate synthetic samples
    newsamples=[]
    for i in range(mult):
        for j in range(data.shape[0]):
            
            #Generate random spot on linear space between data i and one of it's neighbours where
            #to generate the newsample
            alpha=random.uniform(0,1)
            
            #Pick a random neighbour
            nbr_index=random.choice(KNN_indices[j][1:])
            
            #Compute distance for each feature between this sample and the next
            dist=data[j]-data[nbr_index]
            
            #Compute new sample
            newsample=data[j]+alpha*dist
            
            #Store new sample
            newsamples.append(newsample)
            
    return newsamples   
```

We now do some data processing and split the data into test and training samples.

```python
#Normalise Time and Amount variables
df["Time"]=preprocessing.scale(df["Time"].values)
df["Amount"]=preprocessing.scale(df["Amount"].values)

#Split explanatory variables and dependant variables:
df_frauds=df.groupby("Class").get_group(1)
df_frauds_explanatory=df_frauds.iloc[:,:-1]
df_frauds_dependant=df_frauds.iloc[:,-1]

#Split data into test and training sets
X_train_fraud, X_test_fraud, y_train_fraud, y_test_fraud = train_test_split(
    df_frauds_explanatory.values, df_frauds_dependant.values, test_size=0.25, random_state=1)

#Expand training set using SMOTE, generating 2 new samples for every existing one.
mult=2
K=4
X_train_SMOTE=pd.DataFrame(SMOTE(X_train_fraud,mult,K))

#Merge data to get the final enhanced fraud training dataset
X_train_fraud_enhanced=pd.concat([pd.DataFrame(X_train_fraud),X_train_SMOTE])
y_train_fraud_enhanced=np.repeat(y_train_fraud,mult+1)
```

We now undersample from the majority class so the number of observations from the majority class in the training data matches that of the enhances minority class.

```python
#Isolate the legitimate transactions and remove the explanatory variable
df_legit=df.groupby("Class").get_group(0)
df_legit_explanatory=df_legit.iloc[:,:-1]
df_legit_dependant=df_legit.iloc[:,-1]

#Get the enhanced number of total observations
num_fraud=X_train_fraud_enhanced.shape[0]

#Split the majority class into a training dataset equal in size to the enhanced fraud dataset
#and everything else into the test dataset
X_train_legit, X_test_legit, y_train_legit, y_test_legit = train_test_split(
    df_legit_explanatory.values, df_legit_dependant.values, test_size=df_legit.shape[0]-num_fraud, random_state=2)
```

We now merge the datasets and run the logistic regression.

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()
```

```python
#Merge the training data
X_train=pd.concat([pd.DataFrame(X_train_legit),X_train_fraud_enhanced])
y_train=np.concatenate((y_train_legit,y_train_fraud_enhanced))

#Run logit regressions 
reg_array=[0.001,0.01,1,5,10,100]
reg_accuracy=[]
for reg in reg_array:
    logreg=LogisticRegression(penalty='l2',solver="sag",dual=False,C=reg,random_state=1,max_iter=5000)
    kf=KFold(n_splits=10,shuffle=True,random_state=5)
    acc_scores=[]
    for train_index, test_index in kf.split(X_train):
        logreg.fit(X_train.iloc[train_index],y_train[train_index])
        y_pred=logreg.predict(X_train.iloc[test_index])
        acc_sc=accuracy_score(y_train[test_index],y_pred)
        acc_scores.append(acc_sc)
        #cm=confusion_matrix(y_train[test_index],y_pred,labels=[0,1])
    reg_accuracy.append(np.mean(acc_scores))
print(reg_accuracy)
maxindex, maxvalue = max(enumerate(reg_accuracy), key=operator.itemgetter(1))
print("The regularisation term corresponding to the largest accuracy observed in K-fold testing is C=",reg_array[maxindex])
    
```

We therefore set the hyperparameter of C=10. We see that the accuracy steadily grows to that point, after which it seems to stall. This seems to be because the weighted error with a coeffifient of 10 is so much greater than the regularisation term that the latter ends up having virtually no impact on the end result. In this case it seems that regularisation is not helpful.

```python
X_test=pd.concat([pd.DataFrame(X_test_legit),pd.DataFrame(X_test_fraud)])
y_test=np.concatenate((y_test_legit,y_test_fraud))

logreg=LogisticRegression(penalty='l2',solver="sag",dual=False,
                          C=reg_array[maxindex],random_state=1,max_iter=5000)
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
cf=confusion_matrix(y_test,y_pred)
plot_confusion_matrix(cf,["Legitimate","Fraud"])
```
